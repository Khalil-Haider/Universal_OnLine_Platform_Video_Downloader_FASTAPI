```python
import yt_dlp
import os

def download_video(url: str, output_dir: str = ".") -> str:
    ydl_opts = {
        # üéØ Format: best MP4 (merge if needed)
        'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]',
        'merge_output_format': 'mp4',
        'outtmpl': os.path.join(output_dir, '%(title).100s [%(id)s].%(ext)s'),
        
        # üõ°Ô∏è Use deno for JS challenges (critical for YouTube in 2025)
        'js_runtime': 'deno',
        
        # üîç Verbosity (optional)
        'quiet': False,
        'no_warnings': False,
        'noplaylist': True,
        
        # üìÑ Post-processing
        'postprocessors': [{
            'key': 'FFmpegVideoConvertor',
            'preferedformat': 'mp4',
        }],
        
        # üîÑ Retry settings (YouTube loves to 403)
        'retries': 10,
        'fragment_retries': 10,
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=True)
            filename = ydl.prepare_filename(info)
            # If merged, the final file has `.mp4`, not `.webm`
            if not filename.endswith(".mp4"):
                filename = os.path.splitext(filename)[0] + ".mp4"
            print(f"‚úÖ Downloaded: {filename}")
            return filename
    except Exception as e:
        print(f"‚ùå Failed to download {url}: {e}")
        raise

```

```python
url = input("üîó Paste YouTube URL (e.g., https://youtu.be/3MkdEfB_vIo): ").strip()
if url:
    try:
        path = download_video(url)
        print(f"üìÅ Saved at: {os.path.abspath(path)}")
    except Exception:
            print("‚ö†Ô∏è  Aborted.")
```
